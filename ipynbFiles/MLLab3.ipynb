```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

```python
diabetes_X
```

```python
diabetes_y
```

```python
diabetes_X = diabetes_X[:, np.newaxis, 2]
```

```python
diabetes_X
```

```python
#Normalize the Data
diabetes_X = (diabetes_X - np.mean(diabetes_X)) / np.std(diabetes_X)
diabetes_y = (diabetes_y - np.mean(diabetes_y)) / np.std(diabetes_y)
```

```python
#Adding bias
X_b = np.c_[np.ones((diabetes_X.shape[0], 1)), diabetes_X]
```

```python
# Hyperparameters
learning_rate = 0.1
max_iterations = 1000
tolerance = 0.000000001 # Number of samples
```

```python
y = diabetes_y.reshape(-1, 1)
m = len(diabetes_y)
```

```python
#Generating random theta values
theta = np.random.randn(2, 1)
```

```python
# Gradient Descent Loop
cost_history = []
for iteration in range(max_iterations):
    # Compute predictions
    y_pred = X_b.dot(theta)

    # Compute cost function (MSE)
    cost = (1 / m) * np.sum((y_pred - y) ** 2)
    cost_history.append(cost)

    # Compute gradients
    gradients = (2 / m) * X_b.T.dot(y_pred - y)

    # Update parameters
    theta -= learning_rate * gradients

    # Stopping criteria
    if iteration > 0 and abs(cost_history[-1] - cost_history[-2]) < tolerance:
        print(f"Converged at iteration {iteration}")
        break
```

```python
cost_history = []
for iteration in range(max_iterations):
    # Compute predictions
    y_pred = X_b.dot(theta)

    # Compute cost function (MSE)
    cost = (1 / m) * np.sum((y_pred - y) ** 2)
    cost_history.append(cost)

    # Compute gradients
    gradients = (2 / m) * X_b.T.dot(y_pred - y)

    # Update parameters
    theta -= learning_rate * gradients

    # Stopping criteria
    if iteration > 0 and abs(cost_history[-1] - cost_history[-2]) < tolerance:
        print(f"Converged at iteration {iteration}")
        break
```

```python
# Print final parameters and cost
print("Final parameters (theta):\n", theta)
print("Final cost (MSE):", cost_history[-1])
```

```python
# Plot cost function over iterations
plt.plot(range(len(cost_history)), cost_history, label="Cost (MSE)")
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.title("Gradient Descent Progress")
plt.legend()
plt.show()
```

```python
# Plot data and regression line
plt.scatter(diabetes_X, diabetes_y, label="Data")
plt.plot(diabetes_X, X_b.dot(theta), color="red", label="Regression Line")
plt.xlabel("Feature (Normalized BMI)")
plt.ylabel("Target (Normalized y)")
plt.legend()
plt.title("Linear Regression Fit with Gradient Descent")
plt.show()
```
